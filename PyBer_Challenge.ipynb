### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "resources/city_data.csv"
ride_data_to_load = "resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])
#pyber_data_df = ride_data_df.merge(city_data_df, left_on='city', right_on='city')

# Display the data table for preview
pyber_data_df.head()
```

```python
total_drivers = pyber_data_df['driver_count'].sum()
total_drivers 
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
total_rides_by_type = pyber_data_df.groupby(['type']).count()['ride_id']
print(f"Total Rides = {total_rides_by_type}")
urban_total_rides = total_rides_by_type['Urban']
print(f"Urban Total Rides = {urban_total_rides}")
suburban_total_rides = total_rides_by_type['Suburban']
print(f"Suburban Total Rides = {suburban_total_rides}")
rural_total_rides = total_rides_by_type['Rural']
print(f"Rural Total Rides = {rural_total_rides}")
```

```python
# 2. Get the total drivers for each city type
total_drivers_by_type = pyber_data_df.groupby(['type']).sum()['driver_count']
#print(f"Total Drivers = {total_drivers_by_type}")
urban_total_drivers = total_drivers_by_type['Urban']
#print(f"Urban Total Drivers = {urban_total_drivers}")
suburban_total_drivers = total_drivers_by_type['Suburban']
#print(f"Suburban Total Drivers = {suburban_total_drivers}")
rural_total_drivers = total_drivers_by_type['Rural']
#print(f"Rural Total Drivers = {rural_total_drivers}")
total_drivers_by_type
```

```python
# 2.2 Get the total drivers for each city type
total_drivers_by_type = city_data_df.groupby(['type']).sum()['driver_count']

urban_total_drivers = total_drivers_by_type['Urban']
print(f"Urban Total Drivers = {urban_total_drivers}")
suburban_total_drivers = total_drivers_by_type['Suburban']
print(f"Suburban Total Drivers = {suburban_total_drivers}")
rural_total_drivers = total_drivers_by_type['Rural']
print(f"Rural Total Drivers = {rural_total_drivers}")
total_drivers_by_type
```

```python
#  3. Get the total amount of fares for each city type
total_fare_by_type = pyber_data_df.groupby(['type']).sum()['fare']
print(f"Total Fare = {total_fare_by_type}")
urban_total_fare = total_fare_by_type['Urban']
print(f"Urban Total Fare = {urban_total_fare}")
suburban_total_fare = total_fare_by_type['Suburban']
print(f"Suburban Total Fare = {suburban_total_fare}")
rural_total_fare = total_fare_by_type['Rural']
print(f"Rural Total_Fare = {rural_total_fare}")
```

```python
#  4. Get the average fare per ride for each city type. 
avg_fare_per_ride = round(total_fare_by_type/total_rides_by_type, 2)
urban_avg_fare_per_ride = avg_fare_per_ride['Urban']
print(f"Urban average fare per ride percentage = ${urban_avg_fare_per_ride}")
suburban_avg_fare_per_ride = avg_fare_per_ride['Suburban']
print(f"Suburban average fare per ride percentage = ${suburban_avg_fare_per_ride}")
rural_avg_fare_per_ride = avg_fare_per_ride['Rural']
print(f"Rural average fare per ride percentage = ${rural_avg_fare_per_ride}")
avg_fare_per_ride
```

```python
# 5. Get the average fare per driver for each city type. 
avg_fare_per_driver = round(total_fare_by_type/total_drivers_by_type, 2)
urban_avg_fare_per_driver = avg_fare_per_driver['Urban']
print(f"Urban average fare per driver percentage = ${urban_avg_fare_per_driver}")
suburban_avg_fare_per_driver = avg_fare_per_driver['Suburban']
print(f"Suburban average fare per driver percentage = ${suburban_avg_fare_per_driver}")
rural_avg_fare_per_driver = avg_fare_per_driver['Rural']
print(f"Rural average fare per driver percentage = ${rural_avg_fare_per_driver}")
avg_fare_per_driver
```

```python
#  6. Create a PyBer summary DataFrame. 
pyber_summary_df = pd.DataFrame({
    "Total Rides": total_rides_by_type,
    "Total Drivers": total_drivers_by_type,
    "Total Fares": total_fare_by_type,
    "Average Fare per Ride": avg_fare_per_ride,
    "Average Fare per Driver": avg_fare_per_driver})
pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
```

```python
#  8. Format the columns.

pyber_summary_df['Total Rides'] = pyber_summary_df['Total Rides'].map('{:,}'.format)
pyber_summary_df['Total Drivers'] = pyber_summary_df['Total Drivers'].map('{:,}'.format)
pyber_summary_df['Total Fares'] = pyber_summary_df['Total Fares'].map('${:.2f}'.format)
pyber_summary_df['Average Fare per Ride'] = pyber_summary_df['Average Fare per Ride'].map('${:.2f}'.format)
pyber_summary_df['Average Fare per Driver'] = pyber_summary_df['Average Fare per Driver'].map('${:.2f}'.format)
pyber_summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
pyber_data_df.head()
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
pyber_data_by_date_df = pyber_data_df.groupby(['type','date']).sum()[['fare']]
pyber_data_by_date_df
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
# df = df.reset_index()
pyber_data_by_date_df = pyber_data_by_date_df.reset_index()
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
pyber_data_by_date_pivot_df = pyber_data_by_date_df.pivot(index='date', columns='type', values='fare')
pyber_data_by_date_pivot_df
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
pyber_data_by_date_range_pivot_df = pyber_data_by_date_pivot_df.loc['2019-01-01':'2019-04-29']
pyber_data_by_date_range_pivot_df
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
#pyber_data_by_date_pivot_df.index = pd.to_datetime(pyber_data_by_date_pivot_df.index)
pyber_data_by_date_range_pivot_df.index = pd.to_datetime(pyber_data_by_date_range_pivot_df.index)
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
pyber_data_by_date_range_pivot_df.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
pyber_data_by_week_pivot_df = pyber_data_by_date_range_pivot_df.resample("W").sum()
pyber_data_by_week_pivot_df
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 
import matplotlib.dates as mdates
# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')
fig, ax = plt.subplots(figsize=(16, 4))
plt.ylabel('Fare($USD)')
ax.xaxis.set_major_locator(mdates.MonthLocator())
fmt = mdates.DateFormatter('%b')
ax.xaxis.set_major_formatter(fmt)
ax.plot(pyber_data_by_week_pivot_df['Rural'])
ax.plot(pyber_data_by_week_pivot_df['Suburban'])
ax.plot(pyber_data_by_week_pivot_df['Urban'])
ax.set_title("Total Fare by City Type")
mpl.rcParams['legend.loc'] = 'center'
plt.legend(['Rural', 'Suburban', 'Urban'])

plt.show()
```
